```python
import re
import pandas as pd
import numpy as np
```

```python
citationData = pd.read_excel('./新建 Microsoft Excel 工作表.xlsx',header=None)
citationData
```

```python
def extract_article_title(citation):
    # 定义正则表达式以匹配文章标题
    match = re.search(r'\d{4}\.\s(.+?)\.\s', citation)
    if match:
        return match.group(1)
    return None
```

```python
titleList = []
for i in citationData.index:
    tmp = citationData[0][i]
    title = extract_article_title(tmp)
    print(f"Article Title: {title}")
    titleList.append(title)
```

```python
import requests

def get_doi(article_title):
    url = "https://api.crossref.org/works"
    params = {
        "query.bibliographic": article_title,
        "rows": 1
    }
    response = requests.get(url, params=params)
    if response.status_code == 200:
        data = response.json()
        if data["message"]["items"]:
            return data["message"]["items"][0]["DOI"]
        else:
            return "DOI not found."
    else:
        return f"Error: {response.status_code}"

# 示例
doiList = []
with open('./doi.txt','w') as f:
    for i in titleList:
        doi = get_doi(i)
        if i != None:
            print(i +'\t'+doi)
            doiList.append(doi)
            f.write(i +'\t'+doi+'\n')
        else:
            print("error in grep title")
```
