```python
import requests
import re
import os
import pandas as pd
```

```python
kegg = pd.read_excel('./KeggNumber.xlsx',header = None)
kegg.columns = ['number','abbr','SpeciesName']
kegg
```

```python
#species = ['hsa','mcc','ptr','mmu','cel','tup','aml']
species = list(kegg['abbr'])
```

```python
for ID in species:
    url = 'https://metabolicatlas.org/api/v2/gotenzymes/enzymes?filters[organism]='+ID+'&pagination[page]=1&pagination[pageSize]=108500&pagination[column]=gene&pagination[isAscending]=true'
    context = requests.get(url)
    path = './' +ID+'/'
    path1 = './' +ID+'/'+ID+'EnzymeInfo.json'
    path2 = './' +ID+'/'+ID+'EnzymeInfo.txt'
    #check path
    if os.path.exists(path):
        pass
    else:
        os.mkdir(path)
    with open(path1,'w') as f:
        for i in context.text:
            f.write(i)
    tmp = pd.read_json(path1,encoding="utf-8", orient='records')
    tmp = pd.json_normalize(tmp["enzymes"])
    tmp.to_csv(path2,sep='\t')
```
