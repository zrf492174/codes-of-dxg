```python
import pandas as pd  
import matplotlib.pyplot as plt  
import seaborn as sns
import matplotlib as mpl
```

```python
data = pd.read_excel('./data/8gene expression level in brain-mouse.xlsx',sheet_name='Sheet1',header=0)
data['Unnamed: 0'] = data['Unnamed: 0'].str.split('.').str.get(1)
data = data.rename(columns={"Unnamed: 0": "group"})
data.index =data['group']
data.index.name = ''
data
```

```python
std_table = data.groupby(by = 'group').std() # 计算标准差
std_table = std_table.reindex(data['group'].unique().tolist())
figdata = data.groupby(by = 'group').mean()
figdata = figdata.reindex(data['group'].unique().tolist())
```

mpl.rcParams['font.family'] = 'Arial' # 指定默认字体
mpl.rcParams['axes.unicode_minus'] = False

```python
plt.rcParams.update(plt.rcParamsDefault)  
  
```

```python
# 假设figdata和std_table都是pandas的DataFrame  
# 创建一个颜色列表  
colors = plt.colormaps.get_cmap('tab10').colors  # 使用tab10颜色映射，它提供了10种不同的颜色  

# 遍历figdata的列  
for i, col in enumerate(figdata.columns.to_list()):  
    # 选择当前的颜色  
    color = colors[i % len(colors)]  # 如果列的数量超过颜色的数量，则循环使用颜色  
     
    # 绘制errorbar，并添加label  
    plt.errorbar(figdata.index, figdata.loc[:, col], yerr=std_table.loc[:, col],  
                 fmt='-o', color = color,lw=2, ecolor=color, elinewidth=1, ms=7, capsize=3,  
                 label=col)  # 添加label参数  
      
# 添加图例  ,linestyle='--'
plt.legend()  
plt.grid(True, axis='y', color='lightgray', alpha=0.5)
# 在纵轴值为1的位置加一条黑色虚线  
plt.axhline(y=1, color='black', linestyle='--', linewidth=1) 

plt.savefig('./figure/mouseGens.PDF',dpi=300)  
# 显示图形  
plt.show()
```

```python
def errorbarPlot(datapath,filepath):
    data = pd.read_excel(datapath,sheet_name='Sheet1',header=0)
    #data['Unnamed: 0'] = data['Unnamed: 0'].str.split('.').str.get(1)
    data = data.rename(columns={"Unnamed: 0": "group"})
    data.index =data['group']
    data.index.name = ''
    std_table = data.groupby(by = 'group').std() # 计算标准差
    std_table = std_table.reindex(data['group'].unique().tolist())
    figdata = data.groupby(by = 'group').mean()
    figdata = figdata.reindex(data['group'].unique().tolist())
    colors = plt.colormaps.get_cmap('tab10').colors  # 使用tab10颜色映射，它提供了10种不同的颜色  
  
    plt.figure(figsize=(12, 8))
    # 遍历figdata的列  
    for i, col in enumerate(figdata.columns.to_list()):  
        # 选择当前的颜色  
        color = colors[i % len(colors)]  # 如果列的数量超过颜色的数量，则循环使用颜色  

        # 绘制errorbar，并添加label  
        plt.errorbar(figdata.index, figdata.loc[:, col], yerr=std_table.loc[:, col],  
                     fmt='-o', color = color,lw=2, ecolor=color, elinewidth=1, ms=7, capsize=3,  
                     label=col)  # 添加label参数  

    # 添加图例  
    plt.legend() 
    plt.grid(True, axis='y', color='lightgray', alpha=0.5)
# 在纵轴值为1的位置加一条黑色虚线  
    plt.axhline(y=1, color='black', linestyle='--', linewidth=1) 
    plt.xticks(rotation=45)
    plt.savefig(filepath,dpi=300) 
```

```python
datapath = './data/HumanData.xlsx'
filepath = './figure/humanGenes.PDF'
```

```python
errorbarPlot(datapath,filepath)
```

```python
datapath = './data/MouseData.xlsx'
filepath = './figure/mouseGenes.PDF'
errorbarPlot(datapath,filepath)
```
